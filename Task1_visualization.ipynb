```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set(style="whitegrid")
```

```python
url = "https://raw.githubusercontent.com/Prodigy-InfoTech/data-science-datasets/refs/heads/main/Task%201/API_SP.POP.TOTL_DS2_en_csv_v2_38144.csv"
df = pd.read_csv(url, skiprows=4)
```

```python
print("Shape:", df.shape)
display(df.head())
display(df.info())
display(df.describe(include='all').T)
```

```python
# 4.1 Drop any unnamed or metadata columns
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]

# 4.2 Identify year columns (1960-2024) dynamically
year_cols = [col for col in df.columns if col.isdigit()]
df[year_cols] = df[year_cols].apply(pd.to_numeric, errors='coerce')
```

```python
# We will visualise population in year 2020
year = '2020'
data_year = df[year].dropna()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(data_year, bins=20, kde=True)
plt.title(f"Distribution of Population Across Countries ({year})")
plt.xlabel("Population")
plt.ylabel("Number of Countries")
plt.tight_layout()
plt.savefig(f"hist_population_{year}.png", dpi=150)
plt.show()

```

```python
top10 = df[['Country Name', year]].dropna().sort_values(year, ascending=False).head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=year, y='Country Name', data=top10, palette="viridis")
plt.title(f"Top 10 Most Populated Countries ({year})")
plt.xlabel("Population")
plt.ylabel("Country")
plt.tight_layout()
plt.savefig(f"bar_top10_population_{year}.png", dpi=150)
plt.show()
```
